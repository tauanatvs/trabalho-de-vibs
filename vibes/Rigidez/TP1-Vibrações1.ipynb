Dados da máquina MTS - Importação com Pandas

Nota: O primeiro script usava readmatrix() (geralmente para .txt/.csv) e o segundo xlsread().
Ajustei para ler arquivos Excel (.xlsx). Se forem arquivos de texto, mude para pd.read_csv().
Como o MATLAB usa indexação 1 e o Python usa 0, a coluna 3 vira índice 2, e a 5 vira índice 4.


```python
import time
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Lendo o arquivo e forçando a conversão de strings para NaN nas colunas desejadas
df1 = pd.read_excel("test_30psi_20mmmin_Sem EVOL.xlsx")
# K1 = df1.apply(pd.to_numeric, errors='coerce').dropna(subset=[2, 4]).values
K1 = df1.to_numpy()

df2 = pd.read_excel("test_50psi_20mmmin_Sem EVOL.xlsx")
# K2 = df2.apply(pd.to_numeric, errors='coerce').dropna(subset=[2, 4]).values
K2 = df2.to_numpy()

df3 = pd.read_excel("test_60psi_30mmmin_Sem EVOL.xlsx")
# K3 = df3.apply(pd.to_numeric, errors='coerce').dropna(subset=[2, 4]).values
K3 = df3.to_numpy()

df4 = pd.read_excel("test_70psi_30mmmin_Sem EVOL.xlsx")
# K4 = df4.apply(pd.to_numeric, errors='coerce').dropna(subset=[2, 4]).values
K4 = df4.to_numpy()

df5 = pd.read_excel("test_80psi_30mmmin_Sem EVOL.xlsx")
# K5 = df5.apply(pd.to_numeric, errors='coerce').dropna(subset=[2, 4]).values
K5 = df5.to_numpy()


# %% Gráficos de forçaxdeslocamento retirados da máquina MTS
plt.figure(2)
plt.grid(True)

plt.plot(K1[:, 2], K1[:, 4], linewidth=2, color="red", label="30 psi")
plt.plot(K2[:, 2], K2[:, 4], linewidth=2, color="blue", label="50 psi")
plt.plot(
    K3[:, 2], K3[:, 4], linewidth=2, color="#FF12A6", label="60 psi"
)  # Cor aproximada do MATLAB [1 0.07 0.65]
plt.plot(
    K4[:, 2], K4[:, 4], linewidth=2, color="#63D411", label="70 psi"
)  # Cor aproximada do MATLAB [0.39 0.83 0.07]
plt.plot(
    K5[:, 2], K5[:, 4], linewidth=2, color="#9942C7", label="80 psi"
)  # Cor aproximada do MATLAB [0.6 0.26 0.78]

plt.legend(loc="upper left")
plt.axis([0, 100, 0, 7000])
plt.title("Rigidez do Amortecedor Dianteiro", fontweight="bold")
plt.xlabel("Deslocamento (mm)", fontweight="bold")
plt.ylabel("Força (N)", fontweight="bold")

plt.show()
```
